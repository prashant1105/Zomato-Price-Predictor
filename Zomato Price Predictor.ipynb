```python
city = ['New Delhi', 'Gurgaon', 'Noida', 'Faridabad']
```

```python
# Selecting the rows where city is either 'New Delhi', 'Gurgaon', 'Noida', 'Faridabad'...
data = data[data['City'].isin(city)]
```

```python
data.shape
```

```python
# Replacing the city name with some numbers...
data.replace({'New Delhi' : 0, 'Gurgaon' : 1, 'Noida' : 2, 'Faridabad' : 3}, inplace = True)
```

```python
data['Address'].value_counts()
```

```python
# creating a new dataframe which is grouped by Cuisines and mean of Average Cost for two...
x = data.groupby('Cuisines').mean()['Average Cost for two'].reset_index()
x
```

```python
# Merging the dataframe Cuisines and Zomato...
data = data.merge(x, on = 'Cuisines')
```

```python
# Removing the data of y column...
dataX = data.drop('Average Cost for two_x', axis = 1)
dataX.head()
```

```python
X = dataX.iloc[: , :].values
X
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(categorical_features=[0])
```

```python
X = ohe.fit_transform(X).toarray()
```

```python
X.shape
```

```python
X = X[: , 1:]
X.shape
```

```python
ohe1 = OneHotEncoder(categorical_features= [8])
X = ohe1.fit_transform(X).toarray()
```

```python
X.shape
```

```python
X  = np.delete(X, 8, 1)   # It is used to remove a column from somewhere middle of the table... 
# The arguments will be name of the array, index of the array to be removed, row or column (for row 0, column = 1)
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
